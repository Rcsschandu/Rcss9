```python
#scrapping web page
player_name=[]
matches=[]
innings=[]
not_out=[]
runs=[]
highest_score=[]
avg=[]
balls_faced=[]
strike_rate=[]
s100=[]
s50=[]
s4=[]
s6=[]
year=[]
for x in range(2015,2020):
    URL='https://www.iplt20.com/stats/{}/most-runs'.format(x)
    page=requests.get(URL)
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)
    
    

    pname=soup.find_all('div',attrs={'class':'top-players__player-name'})
    for i in pname:
        if i is None:
            player_name.append(np.NaN)
        else:
            player_name.append(i.text)

    match=soup.find_all('td',attrs={'class':'top-players__m top-players__padded'})
    for i in match:
        if i is None:
            matches.append(np.NaN)
        else:
            matches.append(i.text)

    inns=soup.find_all('td',attrs={"class":"top-players__inns"})
    for i in inns:
        if i is None:
            innings.append(np.NaN)
        else:
            innings.append(i.text)


    no=soup.find_all('td',attrs={'class':'top-players__no'})
    for i in no:
        if i is None:
            not_out.append(np.NaN)
        else:
            not_out.append(i.text)


    rs=soup.find_all('td',attrs={'class':'top-players__r is-active'})
    for i in rs:
        if i is None:
            runs.append(np.NaN)
        else:
            runs.append(i.text)


    hs=soup.find_all('td',attrs={'class':'top-players__hs'})
    for i in hs:
        if i is None:
            highest_score.append(np.NaN)
        else:
            highest_score.append(i.text)



    ag=soup.find_all('td',attrs={'class':'top-players__a'})
    for i in ag:
        if i is None:
            avg.append(np.NaN)
        else:
            avg.append(i.text)
             

    bf=soup.find_all('td',attrs={'class':'top-players__b'})
    for i in bf:
        if i is None:
            balls_faced.append(np.NaN)
        else:
            balls_faced.append(i.text)


    sf= soup.find_all('td',attrs={'class':'top-players__sr'})
    for i in sf:
        if i is None:
            strike_rate.append(np.NaN)
        else:
            strike_rate.append(i.text)                               

    hus=  soup.find_all('td',attrs={'class':'top-players__100s'})
    for i in hus:
        if i is None:
            s100.append(np.NaN)
        else:
            s100.append(i.text)

    fs=soup.find_all('td',attrs={'class':'top-players__50s'})
    for i in fs:
        if i is None:
            s50.append(np.NaN)
        else:
            s50.append(i.text)


    fours=soup.find_all('td',attrs={'class':'top-players__4s'})
    for i in fours:
        if i is None:
            s4.append(np.NaN)
        else:
            s4.append(i.text)
    six=soup.find_all('td',attrs={'class':'top-players__6s'})
    for i in six:
        if i is None:
            s6.append(np.NaN)
        else:
            s6.append(i.text)
    for i in range(0,100):
           year.append(x)



```

```python
len(s6)
```

```python
ipl_df=pd.DataFrame({'player':player_name,'no_matches':matches,'no_innings':innings,'not_out':not_out,'no_runs':runs,'high_score':highest_score,'average':avg,'balls_faced':balls_faced,'strike_rate':strike_rate,'no_100s':s100,'no_50s':s50,'no_6s':s6,'no_4s':s4,'season':year})
```

```python
#data cleaning
ipl_df.player=ipl_df.player.apply(lambda x:x.replace('\n',''))
ipl_df.no_matches=ipl_df.no_matches.apply(lambda x:x.replace('\n',''))
ipl_df.no_innings=ipl_df.no_innings.apply(lambda x:x.replace('\n',''))
ipl_df.not_out=ipl_df.not_out.apply(lambda x:x.replace('\n',''))
ipl_df.no_runs=ipl_df.no_runs.apply(lambda x:x.replace('\n',''))
ipl_df.high_score=ipl_df.high_score.apply(lambda x:x.replace('\n',''))
ipl_df.high_score=ipl_df.high_score.apply(lambda x:x.replace('*',''))
ipl_df.average=ipl_df.average.apply(lambda x:x.replace('\n',''))
ipl_df.balls_faced=ipl_df.balls_faced.apply(lambda x:x.replace('\n',''))
ipl_df.strike_rate=ipl_df.strike_rate.apply(lambda x:x.replace('\n',''))
ipl_df.no_100s=ipl_df.no_100s.apply(lambda x:x.replace('\n',''))
ipl_df.no_50s=ipl_df.no_50s.apply(lambda x:x.replace('\n',''))
ipl_df.no_6s=ipl_df.no_6s.apply(lambda x:x.replace('\n',''))
ipl_df.no_4s=ipl_df.no_4s.apply(lambda x:x.replace('\n',''))
ipl_df.head()
```
